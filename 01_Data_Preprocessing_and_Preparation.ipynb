## 8. Weight of Evidence (WoE) and Information Value (IV)

### Manual Calculation First
Before defining the WoE functions, we first understand the underlying calculation manually:

**WoE Formula:**
WoE = ln(Distribution of Good / Distribution of Bad)

**IV Formula:**
IV = Σ (prop_good - prop_bad) × WoE

**IV Interpretation:**
- IV < 0.02: Not predictive — exclude variable
- IV 0.02–0.1: Weak predictor
- IV 0.1–0.3: Medium predictor — include
- IV > 0.3: Strong predictor — include

**WoE Interpretation:**
- Positive WoE: Category has more good loans than average
- Negative WoE: Category has more bad loans than average
- More extreme WoE = stronger separation between good and bad

### WoE Function for Discrete Variables
Encapsulating the manual WoE calculation above into a reusable function that can be applied systematically across all discrete variables. This ensures consistency and efficiency across the full feature set.

```python
def woe_discrete(df, discrete_variabe_name, good_bad_variable_df):
    df = pd.concat([df[discrete_variabe_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    df = df.sort_values(['WoE'])
    df = df.reset_index(drop = True)
    # df['diff_prop_good'] = df['prop_good'].diff().abs()
    # df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```

## Fine Classing — Continuous Variables
Dividing continuous variables into 50 equal-width bins (fine classes) to observe the initial distribution of good and bad loans before coarse classing.

**Why fine classing first:**
With 50 bins we can see the natural pattern of WoE across the full range of each variable before making decisions about which bins to combine in coarse classing.

## Coarse Classing — Combining Bins Based on WoE Monotonicity
Combining fine classes into broader, meaningful categories based on:
1. WoE monotonicity — values should increase or decrease consistently
2. Business logic — categories should make intuitive sense
3. Statistical stability — each category needs sufficient observations

**Why coarse classing matters:**
Too many categories (fine classes) create unstable WoE estimates and overfit the model. Coarse classing reduces noise while preserving the predictive signal of each variable.

### WoE Function for Continuous Variables

```python
def woe_ordered_continuous(df, continuous_variabe_name, good_bad_variable_df):
    df = pd.concat([df[continuous_variabe_name], good_bad_variable_df], axis = 1)
    df = pd.concat([df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].count(),
                    df.groupby(df.columns.values[0], as_index = False)[df.columns.values[1]].mean()], axis = 1)
    df = df.iloc[:, [0, 1, 3]]
    df.columns = [df.columns.values[0], 'n_obs', 'prop_good']
    df['prop_n_obs'] = df['n_obs'] / df['n_obs'].sum()
    df['n_good'] = df['prop_good'] * df['n_obs']
    df['n_bad'] = (1 - df['prop_good']) * df['n_obs']
    df['prop_n_good'] = df['n_good'] / df['n_good'].sum()
    df['prop_n_bad'] = df['n_bad'] / df['n_bad'].sum()
    df['WoE'] = np.log(df['prop_n_good'] / df['prop_n_bad'])
    # df = df.sort_values(['WoE'])
    # df = df.reset_index(drop = True)
    df['diff_prop_good'] = df['prop_good'].diff().abs()
    df['diff_WoE'] = df['WoE'].diff().abs()
    df['IV'] = (df['prop_n_good'] - df['prop_n_bad']) * df['WoE']
    df['IV'] = df['IV'].sum()
    return df
```
